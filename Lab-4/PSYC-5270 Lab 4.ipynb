```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt
from scipy import signal

mat = scipy.io.loadmat('Downloads/chb_sample.mat')
data = mat.get('data')[0]
plt.figure(figsize=(10,10))
plt.subplot(3,1,1)
plt.plot(data[0], 'k-')
```

```python
yref = np.abs(np.fft.rfft(data[0],axis=1))
plt.subplot(3,1,2)
plt.plot(yref, 'k-')
```

```python
plt.subplot(3,1,3)
f, t, Sxx = signal.spectrogram(yref.reshape((921600,)),256)
plt.pcolormesh(t, f, Sxx)
plt.axis((0,3600,0,50))
plt.ylabel('Frequency [Hz]')
plt.xlabel('Time [sec]')
plt.show()
```
