```python
import scipy.io
import numpy as np
import matplotlib.pyplot as plt

mat = scipy.io.loadmat('C:Downloads/nsa2009_1_pyfriendly.mat')

neuron=32
data = mat.get('data')
plt.figure(figsize=(10,7))
plt.subplot(2,1,1)
plt.axis([0,2000,0,data.shape[1]])
for i in range(data.shape[1]-1):
	st = np.squeeze(data[neuron,i])
	plt.plot(st,np.zeros(st.shape)+i,'b.',markersize=2)
plt.axis([0,2000,0,150])

```

```python
binsize = 5
maxTime = 2000
psth = np.zeros([2000,])

st = np.squeeze(data[neuron])

i=1
j=1
for i in range(len(st)):
    c=1
    for j in range(maxTime-1):
            psth[c] = psth[c] + np.sum(np.sum(st[i] >j) + np.sum(st[i]<=(j+binsize)))
            c += 1

pst_time = np.arange(0,2000,1)
plt.subplot(2,1,2)
plt.bar(pst_time,psth-5041);
plt.axis([0,2000,0,100])
plt.show()
```
